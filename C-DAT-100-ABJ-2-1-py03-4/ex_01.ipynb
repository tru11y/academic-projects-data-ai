EXERCISE 01 (5PT)
Display income of “Auvergne-Rhône-Alpes” county from the CVAE enterprise tax for all year (if there is more
than one CVAE income for by year, you must sum all CVAE income from this year)
You must use Pandas dataframe.

```python
import pandas as pd

data=pd.read_csv('dataset.csv')
# dataf=pd.DataFrame(data, columns=['year','reg_name','amount','tax_name'])
result=data[(data['reg_name'].isin(["Auvergne-Rhône-Alpes"]))& (data['tax_name'].isin(["CVAE"]))]
resultat = result.groupby('year')['amount'].sum()
resultat
```

EXERCISE 02 (5PT)
Do the same as exercise 1 but for all regions of France sorted by taxes for all years.
As a example, this is a famous tableau croisé dynamique (pivot table) done by Excel


```python
import pandas as pd
data=pd.read_csv('dataset.csv')
dataf=pd.DataFrame(data, columns=['year','reg_name','amount','tax_name'])
# result=data[data['tax_name']]
# resultat = dataf

dataf.pivot_table(index=['reg_name','year'],values='amount', columns=data.loc[data['tax_name'].isin(["CVAE"]),"tax_name"],aggfunc='sum')
```
