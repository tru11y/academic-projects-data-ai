```python
import pandas as pd
products_df = pd.read_csv('products.csv')
departments_df = pd.read_csv('departments.csv')
# order_products_col=["order_id","product_id","add_to_cart_order"]
# order_products_df = pd.read_csv('order_products.csv', usecols=order_products_col)
columns_type={
    'oder_id':'int32',
    'product_id':'int32',
    'add_to_cart_order':'int16',
    'reordered':'int8',
    "product_name":'string'
}
order_products_df = pd.read_csv('order_products.csv',dtype=columns_type)
orders_df = pd.read_csv('orders.csv')
aisles_df = pd.read_csv('aisles.csv')
# df3.isnull().sum()
order_products_df.duplicated()
# order_products_df
orders_df=orders_df.dropna()
# orders_df
dataf1=pd.merge(products_df, departments_df, on="department_id")
dataf2=pd.merge(dataf1,aisles_df,on="aisle_id")
dataf2

```

# Merging

```python
dataf5=pd.merge(order_products_df,dataf2,on="product_id")
del dataf5['aisle_id']
del dataf5['department_id']
dataf5.to_csv("merge_of_order_products_dataf2.csv")
dataf5
```

# What items sell the most ?

```python
most_sell_items=dataf5["product_name"].value_counts()
most_sell_items=pd.DataFrame(most_sell_items)
most_sell_items=most_sell_items.to_csv("most_sell_items.csv")
# all_product_department.nlargest(8, 'count')
```

# Most selling department

```python
all_product_department=dataf5["department"].value_counts()
all_product_department=pd.DataFrame(all_product_department)
all_product_department.to_csv('all_product_per_department.csv')
all_product_department.nlargest(8, 'count')
```

# Number of products per aisle

```python
all_product_per_aisle=dataf5["aisle"].value_counts()
all_product_per_aisle=pd.DataFrame(all_product_per_aisle)
# all_product_per_aisle.nlargest( 'count')
all_product_per_aisle.to_csv('all_product_per_aisle.csv')
all_product_per_aisle
```

# First product added to cart by customers

```python
first_selected_prod=dataf5.groupby("order_id").first()["product_name"].value_counts()
first_selected_prod=pd.DataFrame(first_selected_prod)
first_selected_prod.to_csv('first_selected_prod.csv')
first_selected_prod.nlargest( 3,'count')
```

# When do people order ?

```python
orders_df = pd.read_csv('orders.csv')
orders_df=orders_df.dropna()
orders_df=pd.DataFrame(orders_df)
days={
    0:"Monday",
    1:"Tuesday",
    2:"Wednesday",
    3:"Thursday",
    4:"Friday",
    5:"Saturday",
    6:"Sunday"
}
orders_df["day"]=orders_df["order_dow"].map(days)
# orders_df=orders_df.groupby(orders_df["day"])
orders_df1=orders_df["day"].value_counts()
orders_df1=pd.DataFrame(orders_df1)
orders_df1.to_csv('order_per_day.csv')
orders_df1
```
