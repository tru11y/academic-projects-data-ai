### You must try and plot all combinaison to determine which features are best to predicting the output. This is called feature selection

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
import pandas as pd

data= pd.read_csv('Fish.csv')
X=data.drop("Species", axis=1)
y = data["Species"]
X_categorical = X.astype(int)
selector = SelectKBest(score_func=chi2, k=2)
X_selected = selector.fit_transform(X_categorical, y)
selected_indices = selector.get_support(indices=True)
print("Selected features:", selected_indices)   
```

# 1 - create a linear regression model using LinearRegression from Scikit-learn

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from  sklearn.metrics import r2_score


data= pd.read_csv('Fish.csv')
data_enc=pd.get_dummies(data, columns=["Species"],dtype=int)
# print(data_enc)
# data["Species"]=data["Species"].astype('category').cat.codes
x=data_enc[['Length1', 'Length2', 'Length3', 'Height', 'Width']]
print("1-LinearRegression model")
y = data_enc["Weight"]
x_train,x_test,y_train, y_test=train_test_split(x, y, test_size=0.2,random_state=42)
model = LinearRegression()
model= model.fit(x_train, y_train )
y_pred=model.predict(x_test)
print(y_pred)
print("2 - plot the test set and the prediction of your model to compare them")
plt.scatter(y_test, y_pred, c='crimson')
plt.plot([y_test.min(), y_test.max()],[y_test.min(),y_test.max()])
plt.title("Linear regression")
plt.legend(["y_true","y_pred"])
plt.show()

print("3-performance indicator")
score= r2_score(y_test, y_pred)
print("R² Score: ",score)
```
